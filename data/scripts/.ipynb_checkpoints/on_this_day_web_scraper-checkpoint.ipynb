```python
import csv 
from datetime import datetime
import requests
import urllib.request
import time
import re
from bs4 import BeautifulSoup
```

```python
months = {1: 'January', 2: 'February', 3:'March', 4:'April', 5:'May', 6:'June', 7:'July', 8:'August', 9:'September', 10:'October', 11:'November', 12:'December'}
file = open('../cleaned_data/on_this_day/dates_cleaned.csv')
reader = csv.reader(file)
garbage = next(reader)
```

```python
# for i in reader:
#     print(datetime.strptime(i[0], '%Y%m%d').strftime('%m/%d/%Y'))
```

```python
with open('../cleaned_data/on_this_day/on_this_day.csv', 'w', newline='\n') as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    writer.writerow(['Date', 'Year', 'Month', 'Day', 'history'])
    
    for i in reader:
        day = datetime.strptime(i[0], '%Y%m%d').day
        month = months[datetime.strptime(i[0], '%Y%m%d').month]
        year = datetime.strptime(i[0], '%Y%m%d').year
        date_string = str(year)  + '_' +  month + '_' + str(day)
        date_pretty = month + ' ' + str(day) + ' ' + str(year)
        print(date_pretty)
        
        # this is to skip until this date since no wikipedia stuff before this date
        if int(i[0]) < 19940701:
            writer.writerow([i[0], year, month, day, ''])
            continue
        else:
            url = 'https://en.wikipedia.org/wiki/Portal:Current_events/'+date_string
            response = requests.get(url)
            soup = BeautifulSoup(response.text, "html.parser")
        
            days_news = re.sub('\[[^<]+?\]', '', re.sub('<[^<]+?>', '', str(soup.find('li')))).replace('\n', '; ').replace('\r', '; ')
            if days_news == 'This page is semi-protected from creation, so only autoconfirmed users can create it.':
                writer.writerow([i[0], year, month, day, ''])
            else:
                print(days_news)
                writer.writerow([i[0], year, month, day, days_news])
            print()

    
```
